# Demo of httpimport

```python
# Must import httpimport!
import httpimport
import pandas as pd
import numpy as np
```

1. Navigate to the file you want to import. Open the raw file.  
![SegmentLocal](images/file-path.gif "file-path")  
2. Copy the url up to but not including the file.py name. Paste it into your jupyter notebook.
3. Pass the name of the desired module to remote_repo.
4. Import the name of the module (should match the exact name passed to remote_repo)

```python
url = 'https://raw.githubusercontent.com/zach-brown-18/class-toolkit/main/eda/'
with httpimport.remote_repo('nan', url):
    import nan
```

You may pass multiple modules within the same directory as a list, as follows:

```python
with httpimport.remote_repo(['nan', 'df_formatting'], url):
    import nan, df_formatting
```
